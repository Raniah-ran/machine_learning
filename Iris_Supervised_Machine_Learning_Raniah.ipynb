# **Preparation of Train Data and Test Data**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(columns=['id', 'species'])
y = df['species']
```

```python
# 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(len(X_train))
print(len(X_test))
```

# **Logistic Regression  96.67%**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Creating and training a Logistic Regression model
model_logistic_regression = LogisticRegression()
model_logistic_regression.fit(X_train, y_train)

# Predict
y_pred = model_logistic_regression.predict(X_test)

# Calculate the accuracy
accuracy_logistic_regression = accuracy_score(y_test, y_pred)
print(f'Accuracy of Logistic Regression Model: {accuracy_logistic_regression * 100:.2f}%\n')

print('Classification Report:')
print(classification_report(y_test, y_pred))
```

# **K-NN  100%, 96.67%**

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Search the optimal k value
k_range = range(1, 26)
scores = []

for k in k_range:
    model_search_k = KNeighborsClassifier(n_neighbors=k)
    cv_scores = cross_val_score(model_search_k, X_train, y_train, cv=5)
    mean_score = cv_scores.mean()
    scores.append(mean_score)

    print(f'K = {k}: Mean Accuracy = {mean_score * 100:.2f}%')

plt.plot(k_range, scores)
plt.xlabel('K value')
plt.ylabel('Mean Accuracy')
plt.title('K Value Selection in K-NN')
plt.show()

best_k = k_range[scores.index(max(scores))]
best_mean_accuracy = max(scores)

print(f'Optimal K: {best_k}, with the mean accuracy of {best_mean_accuracy * 100:.2f}%')
```

even though the optimal K is 11, let's try to see the accuracy of K-NN with K = 3 and 11

```python
k_values = [3, 11]
results = []

for k_value in k_values:
    model_KNN = KNeighborsClassifier(n_neighbors=k_value)
    model_KNN.fit(X_train, y_train)

    # Predict
    y_pred = model_KNN.predict(X_test)

    # Calculate the accuracy
    accuracy_KNN = accuracy_score(y_test, y_pred)
    print(f'K = {k_value}: Accuracy of KNN model = {accuracy_KNN * 100:.2f}%\n')

    results.append((k_value, accuracy_KNN, y_pred))

    print('Classification Report:')
    print(classification_report(y_test, y_pred))
```



with k = 3, the accuracy is 100% but:
*  The model may overfit the training data, causing it to perform poorly on new, unseen data, especially if the new data is a little different from the training data.
*   We might only be able to use this model on small data.

with k = 11, the accuracy is 96.67%
*   A larger k value, like 11, tends to be more stable and generalizable. This means the model doesn't overfit the training data and can recognize general patterns that work on new data
*   more suitable for real-world use with more varied data.



# **Decision Tree  96.67%**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
# Search the optimal max_depth
max_depth_range = range(1, 11)
scores = []

for depth in max_depth_range:
    model_tree = DecisionTreeClassifier(criterion='gini', max_depth=depth, random_state=1)
    cv_score = cross_val_score(model_tree, X_train, y_train, cv=5)
    mean_score = cv_score.mean()
    scores.append(mean_score)

    print(f'depth = {depth}: Mean Accuracy = {mean_score * 100:.2f}%')

plt.figure(figsize=(6, 4))
plt.plot(max_depth_range, scores, marker='o', linestyle='--', color='b')
plt.xlabel('Max Depth')
plt.ylabel('Mean Accuracy')
plt.title('Max Depth Selection in Decision Tree')
plt.grid()
plt.show()

best_depth = max_depth_range[scores.index(max(scores))]
best_mean_accuracy = max(scores)
print(f'Optimal Max Depth: {best_depth}, with the mean accuracy of {best_mean_accuracy * 100:.2f}%')
```

```python
# Create and train a Decision Tree model
model_tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=1)
model_tree.fit(X_train, y_train)

# Predict
y_pred = model_tree.predict(X_test)

# Calculate the accuracy
accuracy_tree = accuracy_score(y_test, y_pred)
print(f'Accuracy of Decision Tree Model: {accuracy_tree * 100:.2f}%\n')

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Decision Tree Visualization
plt.figure(figsize=(12, 8))
plot_tree(model_tree,
          feature_names=['sepallengthcm', 'sepalwidthcm', 'petallengthcm', 'petalwidthcm'],
          class_names=model_tree.classes_,  # Target class name
          filled=True,  # Node colors based on class
          rounded=True,  # Create a round node
          fontsize=10)
plt.title("Decision Tree Visualization")
plt.show()
```

# **Random Forest  96.67%**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Search the optimal n_estimators_range
n_estimators_range = range(10, 201, 10)  # Test from 10 to 200 trees, with a range of 10 from one value to another
scores = []

for n in n_estimators_range:
    model = RandomForestClassifier(n_estimators=n, random_state=1)
    cv_score = cross_val_score(model, X_train, y_train, cv=5).mean()
    scores.append(cv_score)

plt.plot(n_estimators_range, scores)
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Mean Accuracy')
plt.title('n_estimators Selection for Random Forest')
plt.show()

best_n_estimators = n_estimators_range[scores.index(max(scores))]
best_mean_accuracy = max(scores)
print(f'n_estimators Optimal: {best_n_estimators}, with the mean accuracy of {best_mean_accuracy * 100:.2f}%')
```

```python
# Create and train a Random Forest model
model_rf = RandomForestClassifier(n_estimators=10, random_state=1)
model_rf.fit(X_train, y_train)

# Predict
y_pred = model_rf.predict(X_test)

# Calculate the accuracy
accuracy_rf = accuracy_score(y_test, y_pred)
print(f'Accuracy of Random Forest Model: {accuracy_rf * 100:.2f}%\n')

print("Classification Report:")
print(classification_report(y_test, y_pred))
```
